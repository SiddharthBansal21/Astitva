```python
import re

def extract_pan_info(json_data):
    pan_number = None
    name = None
    fathers_name = None
    dob = None

    # Loop through the pages
    for page in json_data.get('pages', []):
        # Loop through blocks within a page
        for block in page.get('blocks', []):
            for line in block.get('lines', []):
                line_text = ' '.join([word['value'] for word in line.get('words', [])])

                # Extract PAN Number (10 character alphanumeric)
                pan_match = re.search(r'\b[A-Z]{5}[0-9]{4}[A-Z]{1}\b', line_text)
                if pan_match:
                    pan_number = pan_match.group(0)

                # Extract Name (assuming it's in all caps)
                if re.match(r'^[A-Z\s]+$', line_text) and len(line_text) > 5:
                    if not name:
                        name = line_text.strip()
                    elif not fathers_name:
                        fathers_name = line_text.strip()

                # Extract Date of Birth
                dob_match = re.search(r'\b\d{2}/\d{2}/\d{4}\b', line_text)
                if dob_match:
                    dob = dob_match.group(0)

    return {
        'PAN Number': pan_number,
        'Name': name,
        'Father\'s Name': fathers_name,
        'Date of Birth': dob
    }

# Example usage:
json_response = {
    'pages': [{'blocks': [
        # ... (your provided JSON data here)
    ]}]
}

result = extract_pan_info(json_response)
print(result)
```
